```python
# Goal: Try out the different plotting packages: 
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import image as mpimg # I think image is matplotlib's image functionality module?
import numpy as np
import cv2
```

```python
img_mp = mpimg.imread('RVIZ_Test_Image#6_Grass.png',0)

img_cv = cv2.imread('RVIZ_Test_Image#6_Grass.png', 0)
```

```python
# Display the two images: 
print img_mp.shape
plt.imshow(img_mp)
```

```python
print img_cv.shape
#plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
plt.imshow(img_cv, cmap ='gray')
```

```python
fig , axes = plt.subplots(1,2, sharex = True)
print len(axes)

axes[0].imshow(img_mp)
axes[1].imshow(img_cv, cmap = 'gray')
axes[0].set_title('MP plot')
#axes[1].title= 'CV_PLOT'
```

```python
# A bit more basic plotting: 
# Simple data to display in various forms
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig2 , ax = plt.subplots()

fig2.suptitle("Figure Title")
ax.plot(x, y)
ax.grid()
```
