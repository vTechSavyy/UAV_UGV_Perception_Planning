```python
# Import libraries:
import numpy as np
import time
from A_star_Dijkstra_utilities import *
import importlib
from matplotlib import pyplot as plt

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 


import params

importlib.reload(params)

# Generate grid and return X and Y co-ordinates of all the nodes:
[params.X,params.Y] = generate_grid()
```

```python
## -2.a Run the Dijkstra algorithm - Using a Python Dictionary for the Priority Queue: 
# Setup the algorithm:
# length ~ 130

numNodes = len(params.X)

# Define the start and goal nodes:
startIdx = get_node_idx(0,0)
goalIdx =  get_node_idx(30.0,10.0)
epsilon = 0.0  # This makes A* essentially Dijkstra

start = time.time()

# Run the Dijkstra algorithm:
dijDictPathIndices , dictLengthDij = astar_dict(startIdx,goalIdx, numNodes, epsilon)  

end = time.time()


timeDijDict = end - start

print (" Time taken for Dijkstra algorithm using a Python Dictionary is : " , timeDijDict ,  "secs")
print (" The length of the Path is: " , dictLengthDij)
print (" The Grid Resolution is: " , params.gridRes)
```
